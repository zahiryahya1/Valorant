# Testing queries, visualizations, experimenting

```python
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY=os.getenv("HENRIKDEV_API_KEY")
```

```python

print("Loaded key:", API_KEY)
```

```python
import requests

name = "Captain Ignant"
tag = "goat" 

headers = {
    "Authorization": API_KEY
}

url = f"https://api.henrikdev.xyz/valorant/v2/account/{name}/{tag}"

response = requests.get(url, headers = headers)

print("Status:", response.status_code)
print(response.json())
```

```python
import json

account_data = response.json()
puuid = account_data.get("data", {}).get("puuid")
region = account_data.get("data", {}).get("region")


matchs_url = f"https://api.henrikdev.xyz/valorant/v3/by-puuid/matches/{region}/{puuid}"

response = requests.get(matchs_url, headers=headers)

print("Status:", response.status_code)

with open("matches.json", "w") as f:
    json.dump(response.json(), f, indent=4)
```

```python
match_id = "9ad1bb4b-e17a-4dd1-9489-ae11b4ed1fa6"
matchurl = f"/valorant/v2/match/{match_id}"

response = requests.get(f"https://api.henrikdev.xyz{matchurl}", headers=headers)
print(response.json())
with open("single_match.json", "w") as f:
    json.dump(response.json(), f, indent=4)
```
